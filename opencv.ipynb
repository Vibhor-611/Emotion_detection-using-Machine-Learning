Blurred

```python
img1=cv2.imread("C:/Users/Vibhor Rajput/Desktop/emtotion_detection_course/archive/images/7/Happy.jpg")
img=cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
figure_size=7
result=cv2.blur(img,(figure_size,figure_size))
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(cv2.cvtColor(img1,cv2.COLOR_HSV2RGB)),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(result,cv2.COLOR_HSV2RGB)),plt.title("Blurred")
plt.xticks([]),plt.yticks([])
plt.show()
```
